# ลบ Missing

missing -> null, na (not autorized), NaN (not a number)

```python
unit.dropna()
```

## axis{0 or ‘index’, 1 or ‘columns’}, default 0

axis = 0 : Drop rows which contain missing value

axis = 1 : Drop columns which contain missing value

```python
unit.dropna(axis=1) # axis = 1 : Drop columns which contain missing value
```

## how{‘any’, ‘all’}, default ‘any’

‘any’ : If any NA values are present, drop that row or column.

‘all’ : If all values are NA, drop that row or column.

```python
unit.dropna(how='all')
```

## thresh : int, optional
Require that many non-NA values.

```python
unit.dropna(axis=1,thresh=42000)
```

```python
unit.head().dropna(axis=0, thresh=10)
```

## subset : column label or sequence of labels, optional
Labels along other axis to consider, e.g. if you are dropping rows these would be a list of columns to include.

ดูแค่คอลัมน์ที่เรากำหนด

```python
unit.dropna(subset=['propertytype_name_en','price_min','area_usable_min'])
```

```python
unit1 = unit.dropna(subset=['propertytype_name_en','price_min','area_usable_min'])
```

## inplace : bool, default False
If True, do operation inplace and return None.

## 1.เปรียบเทียบราคาเฉลี่ยของ Detached house, Condo, Twin house และ Townhome 

ตอบ 6.674175e+06, 6.266300e+06, 3.612469e+06, 2.810930e+06

```python
unit.groupby(['propertytype_name_en'])['price_min'].mean().filter(regex='(Detached House|Condo|Twin house|Townhome)')
```

```python
unit.groupby(['propertytype_name_en']).mean()[['price_min']]
```

## 2.พื้นที่ (area_total_min_wa และ area_usable_min) และ จำนวนห้องนอนของทรัพย์สินที่แพงที่สุดแต่ละชนิด



```python
unit.groupby(['propertytype_name_en'])['price_min'].max()
```

```python
unit.groupby(['propertytype_name_en']).max()[['price_min','area_total_min_wa','area_usable_min','count_room_bed']]
```

```python
unit[unit['price_min']== unit['price_min'].max()][['propertytype_name_en','price_min','area_total_min_wa','area_usable_min','count_room_bed']]
```

## 3.จำนวนที่จอดรถและห้องน้ำของทรัพย์สินที่ถูกที่สุดแต่ละชนิด



```python
unit.groupby(['propertytype_name_en'])['price_min'].min()
```

```python
unit.groupby(['propertytype_name_en']).min()[['price_min','count_parking','count_room_bath']]
```

```python
unit[unit['price_min']== unit['price_min'].min()][['propertytype_name_en','price_min','area_total_min_wa','area_usable_min','count_room_bed']]
```
